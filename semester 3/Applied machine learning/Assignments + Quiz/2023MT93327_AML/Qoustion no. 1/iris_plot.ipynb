importing packages and loading data

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import preprocessing
import seaborn as sns

iris = pd.read_csv("iris-3.csv")
print(iris.head())
```

Compute mean and standard deviation

```python

numeric_attributes = iris.select_dtypes(include='number')
mean_values = numeric_attributes.mean()
std_dev_values = numeric_attributes.std()

print("Mean values:\n", mean_values)
print("\nStandard deviation values:\n", std_dev_values)
```

Converting string labels of the species' column into numbers.

```python
le = preprocessing.LabelEncoder()
iris.species = le.fit_transform(iris.species)
print(iris.head())
```

Creating a scatterplot (A) Sepal lenght VS Petal Length

```python
sns.scatterplot(data=iris, x='sepal_length',
                y='petal_length', hue='species')
plt.plot()
```

Creating a scatterplot (B) Petal Width VS Sepal Width

```python
sns.scatterplot(data=iris, x='petal_width',
                y='sepal_width', hue='species')
plt.plot()
```

Creating a scatterplot (C) Petal Length VS Sepal Length

```python
sns.scatterplot(data=iris, x='petal_length',
                y='sepal_length', hue='species')
plt.plot()
```

Visualize 5 point summary

```python
five_point_summary = numeric_attributes.describe().transpose()
plt.figure(figsize=(10, 6))
sns.boxplot(data=five_point_summary, orient='h')
plt.title('5-Point Summary for Numeric Attributes')
plt.show()
```

Create histograms with 5 bins for Sepal Length and Petal Width

```python
plt.figure(figsize=(10, 6))
plt.subplot(1, 2, 1)
plt.hist(iris['sepal_length'], bins=5, edgecolor='black')
plt.xlabel('Sepal Length')
plt.ylabel('Frequency')
plt.title('Histogram of Sepal Length')
plt.subplot(1, 2, 2)
plt.hist(iris['petal_width'], bins=5, edgecolor='black')
plt.xlabel('Petal Width')
plt.ylabel('Frequency')
plt.title('Histogram of Petal Width')
plt.tight_layout()
plt.show()
```
